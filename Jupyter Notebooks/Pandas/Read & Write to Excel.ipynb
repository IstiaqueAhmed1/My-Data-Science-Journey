# Converters argument in 'read_excel()' method. 

```python
import pandas as pd
def convert_people_cell(cell):
    if cell=='n.a.':
        return 'Sam Walton'
    else:
        return cell
def convert_eps_cell(cell):
    if cell=='not available' or cell=='n.a.':
        return None
    else:
        return cell
def convert_price_cell(cell):
    if cell=='not available' or cell=='n.a.':
        return None
    else:
        return cell
df=pd.read_excel(r'C:\Users\h\Documents\GitHub\py\pandas\4_read_write_to_excel\stock_data.xlsx','Sheet1',converters={
    'people':convert_people_cell,
    'eps':convert_eps_cell,
    'price':convert_price_cell
})


df
```

# Write two dataframes to two separate sheets in excel

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter(r'C:\Users\h\Documents\GitHub\py\pandas\4_read_write_to_excel\stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer,sheet_name="Stocks",index=False)
    df_weather.to_excel(writer,sheet_name="Weather",index=False)

```
