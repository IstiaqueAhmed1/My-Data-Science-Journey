# Replace invalid values to Nan Value with 'replace()' function

```python
import pandas as pd
import numpy as np
df=pd.read_csv('weather_data.csv')
df
```

```python
new_df=df.replace(-99999,np.NaN)              #Insert the (value,np.NaN) into the replace function to replae the value with an NaN value.
new_df
```

```python
new_df=df.replace([-99999,-88888],np.NaN)    #To replace multiple invalid values; put a list of those invalid values as an argument to the 'replace()' function.
new_df
```

```python
new_df=df.replace({                    #Insert a dictionary with column keys ans invalid values to replace them with NaN value.
    'temperature':-99999,
    'windspeed':[-99999,-88888],
    'event':'0'
},np.NaN)

new_df
```

```python
new_df=df.replace({
    -99999:np.NaN,
    -88888:np.NaN,
    '0':'Sunny'
})
new_df
```

#  Replace with regular expressions

```python
new_df=df.replace({
    'temperature':['A-Za-z'],               #Using regular expression on selected columns as a dictionary to replace alphabets with blank space.
    'windspeed':['A-Za-z'],
},'',regex=True)

new_df
```

# Replace list of values with another list of values
