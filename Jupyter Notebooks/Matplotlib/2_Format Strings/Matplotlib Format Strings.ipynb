# Matplotlib Formatting Strings with format argument

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,"c*--")                           #The 3rd argument is for formatting a plot
```

```python
plt.plot(x,y,color='#000000',marker='D',linestyle='',linewidth=3,ms=7)   #'marker=' argument is for selecting the marker
                                                                         #'markersize=' or 'ms=' is to selct the size of the marker
```

```python
plt.plot(x,y,alpha=0.5)                                  #'alpha=' is for selecting the thranparancy of the line in the plot.
```
