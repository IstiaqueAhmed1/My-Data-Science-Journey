```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]    
```

```python
plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')
plt.hist(blood_sugar,bins=3,rwidth=0.95)                  #use 'bins' to select in how many sections you want your data to be divided                            
                                                          #use 'rwidth' to create the gap between the sections
                   
```

```python
plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.9,color='pink')      #create a list of the exact range you divide the data into and put it into the 'bins=' method.
```

# Creating side by side histograms

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.9,color=['#0B0B45','pink'],label=["Men",'Women'],orientation='horizontal')
plt.legend(loc='best',shadow=True,fontsize='x-large')
```
