# Creating a Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis('equal')                                   #To make the circle perfect.
plt.pie(exp_vals,labels=exp_labels,radius=1.5)      #Use 'plt.pie()' to create a pie chart. #Use'radius=' to set the size of the pie chart.
plt.show()                                          #It is used to show the the chart. It is a must if you are not using a Jupyter Notebook.
                                                    
```

# Creating Pie Chart with Percantages

```python
plt.axis('equal')                                                                                                      #Use "'autopct='%f%%'" to set percentages to each values.
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,0,0.2,0,0.2],startangle=90)     #Use 'startangle=' to set the angle of the first value.
plt.show()                                                                                                             #Use'explode=' to take any value out of the pie chart.
```
