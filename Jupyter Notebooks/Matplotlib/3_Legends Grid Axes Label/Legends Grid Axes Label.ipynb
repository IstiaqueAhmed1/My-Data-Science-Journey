```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel(' Temperature')
plt.title('Weather')

plt.plot(days,max_t,label="Max")          #'label=' is for labeling each lines in the plot
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc='best',shadow=True,fontsize='x-large')            #'plt.legend(loc=)' is for showing the labels of the lines in the plot
                                                                 #'shadow=' is for creating a shadow around the legend box
                                                                 #'fontsize=' is for select the size of the font of the legend box
plt.grid()                                                       #'plt.grid()' is for creating gridlines in the plot
```
